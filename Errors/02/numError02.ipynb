```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
def f(x: int) -> float:
    return ((x-1)**6)
def q(x: int) -> float:
    return ((x**6) - (6*(x**5)) + 15*(x**4) - 20*(x**3) + 15*(x**2) - (6*x) + 1)
```

```python
rangeAn = [4.8,44.2]
```

```python
x1 = np.linspace(rangeAn[0], rangeAn[1], 5000)
y1 = f(x)

x2 = np.linspace(rangeAn[0], rangeAn[1], 5000)
y2 = q(x)
```

```python
fig = px.scatter(title='Function p(x)=(x−1)^6 and q(x)=x^6−6x^5+15x^4−20x^3+15x^2−6x+1')
fig.update_layout(width=1600, height=700)
fig.add_scatter(x=x2, y=y2, mode='lines', name='q(x)')
fig.add_scatter(x=x1, y=y1, mode='lines', name='p(x)')
fig.show()
```
