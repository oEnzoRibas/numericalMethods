```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd 
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.io as pio
import plotly.subplots as sp
import plotly.express as px
```

```python
def f(x: int) -> float:
    return (x**10) + 1 - (x**10)
```

```python
x = np.linspace(-60, 60, 5000)
y = f(x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function f(x) = (x^10) + (1) − (x^10)')
plt.gcf().set_size_inches(22, 6)
plt.grid(True)
```

```python
x = np.linspace(-60, 60, 5000)
```

```python
fig = px.line(x=x, y=y, labels={'x': 'x', 'y': 'f(x)'}, title='Function f(x) = (x^10) + (1) − (x^10)')
fig.update_layout(hovermode="x unified")
fig.show()
```

# Base Function


```python
def i(x: int, l: np.float32) -> float:
    return (x**10) +l - (x**10)
```

```python
literal = np.float32(12)

x = np.linspace(-145, 142, 5000)
y = i(x,literal)

fig = px.line(x=x, y=y, labels={'x': 'x', 'y': 'f(x)'}, title=f'Function f(x) = (x^10) + {literal} − (x^10)')
fig.update_layout(hovermode="x unified")
fig.show()
```

```python
def funcaoTeste(x: int, l: np.float32) -> float:
    return (x**10) +l - (x**10)
```

```python
literal = np.float32(12)

x = np.linspace(-145, 142, 50200)
y = funcaoTeste(x,literal)

fig = px.line(x=x, y=y, labels={'x': 'x', 'y': 'f(x)'}, title=f'Function f(x) = (x^10) + {literal} − (x^10)')
fig.update_layout(hovermode="x unified")
fig.show()
```

```python
fig = go.Figure()

x = np.linspace(-145, 142, 5000)
y = funcaoTeste(x,12)


# First graph
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Graph 1', line=dict(color='blue')))

# Second graph (example with a shifted y for demonstration)
y2 = funcaoTeste(x,5)

fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='Graph 2', line=dict(color='red')))

fig.update_layout(
    title='Two Graphs with Different Colors',
    xaxis_title='x',
    yaxis_title='f(x)',
    hovermode="x unified"
)

fig.show()
```

```python
fig = go.Figure()

x = np.linspace(-145, 142, 5000)
y = funcaoTeste(x,1)

for i in range(1, 15):
    y = funcaoTeste(x, i)
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name=f'Graph {i}'))

fig.update_layout(
    title='Graphs with Different Colors',
    xaxis_title='x',
    yaxis_title='f(x)',
    hovermode="x unified"
)

fig.show()
```
