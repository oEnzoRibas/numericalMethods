```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
def f(x: int) -> float:
    return (x**10) + 1 - (x**10)
```

```python
x = np.linspace(-60, 60, 5000)
y = f(x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function f(x) = (x^10) + (1) âˆ’ (x^10)')
plt.gcf().set_size_inches(22, 6)
plt.grid(True)
```

```python
x = np.linspace(-60, 60, 5000)
```

```python
fig = px.line(x=x, y=y, labels={'x': 'x', 'y': 'f(x)'}, title='Function f(x) = (x^10) + (1) âˆ’ (x^10)')
fig.update_layout(hovermode="x unified")
fig.show()
```
